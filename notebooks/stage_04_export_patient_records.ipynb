# Stage 4 — Export Patient Records

Writes per-patient folders with clinical notes, ground truth, information extraction, and symptom trees.

**Input:** Stage 1 cohort + Stage 3 symptom tree results  
**Output:** `patient_records/`


## Setup & Export

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if (ROOT / "pipeline.py").exists():
    NB_DIR = ROOT
elif (ROOT / "notebooks" / "pipeline.py").exists():
    NB_DIR = ROOT / "notebooks"
else:
    NB_DIR = ROOT.parent / "notebooks"
sys.path.insert(0, str(NB_DIR))

from pipeline import (
    EXPORT_DIR,
    export_cohort_to_folders,
    load_cohort,
    load_symptom_tree_results,
    print_pipeline_banner,
)

print_pipeline_banner()
cohort_df = load_cohort()
results_df, patient_symptom_trees = load_symptom_tree_results()
export_root = export_cohort_to_folders(
    cohort_df, results_df, output_dir=EXPORT_DIR, patient_symptom_trees=patient_symptom_trees
)
print(f"Exported to: {export_root.resolve()}")
```
