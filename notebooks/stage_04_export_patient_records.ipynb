## Setup & Export

```python
import sys
from pathlib import Path

REPO_ROOT = Path.cwd()
if not (REPO_ROOT / "export_patient_data.py").exists():
    REPO_ROOT = Path.cwd().parent
sys.path.insert(0, str(REPO_ROOT))

from cohort_selection import load_cohort
from export_patient_data import export_cohort_to_folders
from pipeline_config import EXPORT_DIR, print_pipeline_banner
print_pipeline_banner()

from pipeline_io import load_symptom_tree_results

cohort_df = load_cohort()
results_df, patient_symptom_trees = load_symptom_tree_results()

export_root = export_cohort_to_folders(
    cohort_df,
    results_df,
    output_dir=EXPORT_DIR,
    patient_symptom_trees=patient_symptom_trees,
)
print(f"Exported to: {export_root.resolve()}")
```
